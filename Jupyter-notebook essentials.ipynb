# Keyboard shortcuts

esc- command mode

enter-again back to edit mode

in command mode you can lot of operations:
a:create cell above
b: create cell below
arrow :to go above or below cells
shift arrows: to select multiple cells
dd:delete cell
m:markdown
y:code mode
shift +enter:execute cell
select cells then shift+m:to merge cells
control+shift+minus:split a cell at certain point(be in that point first)
tab :auto complete
shift+tab inside a function:details of that method /function
